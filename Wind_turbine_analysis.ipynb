# Data Pre-Processing

# Determing null/missing  values

```python
num_col.columns
```

```python
num_col.isna().sum()
```

# Removing missing values

```python
from sklearn.impute import SimpleImputer
mean_imputer = SimpleImputer(strategy='mean')
num_col = pd.DataFrame(
    mean_imputer.fit_transform(num_col), columns=['Wind_speed', 'Power', 'Nacelle_ambient_temperature',
       'Generator_bearing_temperature', 'Gear_oil_temperature',
       'Ambient_temperature', 'Rotor_Speed', 'Nacelle_temperature',
       'Bearing_temperature', 'Generator_speed', 'Yaw_angle', 'Wind_direction',
       'Wheel_hub_temperature', 'Gear_box_inlet_temperature'])
    
print(num_col.isna().sum())
```

# Determining  duplicates 

```python
num_col.columns
```

```python
num_col.duplicated(keep="first")

```

# To determine  and removing outliers

```python
import numpy as np

for col in num_col.columns:
        print(col)
        Q1 = num_col[col].quantile(0.25)
        Q3 = num_col[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        num_col[col] = pd.DataFrame(np.where(num_col[col] > upper_bound, upper_bound, np.where(num_col[col] < lower_bound, lower_bound,
                                                                                         num_col[col])))    

print(num_col)
```

```python
sns.set(rc={'figure.figsize':(25, 15)})
sns.boxplot(data=num_col)

  
```

# To determine the skewness of the columns to apply manual transformation

```python
num_col.skew()
```

# Applying Log transformation to all columns with skewness out of range from -0.5 to 0.5  except power column

```python
import scipy.stats as stats
import pylab
import numpy as np





# Assuming your DataFrame is named df

# Columns with moderate positive skewness
sqrt_cols = [
    'Wind_speed',
    'Gear_oil_temperature',Applying Log transformation to all columns with skewness out of range from -0.5 to 0.5 except power column
    'Ambient_temperature',
    'Rotor_Speed',
    'Generator_speed'
]

# Apply square root transformation
for col in sqrt_cols:
    # Add +1 to handle any zero values safely
    num_col[col + '_sqrt'] = np.sqrt(num_col[col] + 1)



# Check new skewness values after transformation
print("After Transformation:", num_col[[col + '_sqrt' for col in sqrt_cols]].skew())
print("Before transformation:", num_col[[col  for col in sqrt_cols]].skew())
```

# Applying yeo johnson transformation to power column

```python
from feature_engine import transformation

# Set up the Yeo-Johnson transformer for 'workex' variable
tf = transformation.YeoJohnsonTransformer(variables = 'Power')

# Transforming the 'workex' variable using Yeo-Johnson transformation
num_col_tf= pd.DataFrame(tf.fit_transform(num_col),columns= num_col.columns.tolist())
num_col_tf
# Transformed data
# Checking whether the transformed 'workex' data is normally distributed using a Q-Q plot
prob = stats.probplot(num_col_tf.Power, dist = stats.norm, plot = pylab)
pylab.title("Q-Q Plot after Yeo-Johnson Transformation on Power")
pylab.show()
```

# As there is no outliers in dataset. we go with standard scaler method for normal distrubution

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy import stats
import matplotlib.pylab as pylab

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform automatically, keeping same column names
num_col_ss = pd.DataFrame(
    scaler.fit_transform(num_col),
    columns=num_col.columns  # 
)

# Verify dimensions match
print("Scaled shape:", num_col_ss.shape)
print("Original shape:", num_col.shape)

# Plot Q-Q plot for 'Rotor_Speed'
stats.probplot(num_col_ss['Rotor_Speed'], dist=stats.norm, plot=pylab)
pylab.title("Q-Q Plot after Standard Scaling on Rotor_Speed")
pylab.show()

stats.probplot(num_col_ss['Gear_oil_temperature'], dist=stats.norm, plot=pylab)
pylab.title("Q-Q Plot after Standard Scaling on Yaw_angle ")
pylab.show()
```
